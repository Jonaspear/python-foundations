## List vs. Arrays

1. A Python list is a storage for values, NumPy arrays are also a storage but have the ability to comprehend mathimatical concepts.

2. Shape simpily refers to how the data is organized and can be understood by looking at the rows and columns, (4, 2) would refer to 4 rows and 2 columns, rows are always first.

```python
import numpy as np

grades = np.array([67, 89, 91, 96, 99])

print("Array:", grades)

print("Shape:", grades.shape)
```

## 1D array vs. 2D Array

1. Looking up above I assumed the output refered to 5 rows and 0 columns but in actuallity a number with no folowing output after the comma means it has only one dimension which also means for the example above and those like it, it is impossible for it to have a row or column.

2. A 1D array is a line of numbers or in other words a 1 dimensional line of numbers, a 2D array which is often seen as a grid only appears when both rows and columns are present.

```python
import numpy as np 

student_scores = np.array ([
    [21, 77, 44],
    [67, 56, 98],
    [20, 12, 35]
])

print("Array:", student_scores)

print("Shape:", student_scores.shape)
```

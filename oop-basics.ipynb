# OOP (Object Oriented Programming) Basics

## Introduction

1. A class is like a template, it is the overarching formation on how the information is formatted.

2. An object is an instance of the template using unique data/information.

3. An example would be recording someones KDA within a video game, the format of K/D/A would be the class, the object would be the unique information each player would have like Player 1 is going 5/1/2.

```python
class Player:
    def __init__(self, name, kills, deaths, assists):
        self.name = name
        self.kills = kills
        self.deaths = deaths
        self.assists = assists
        
player1 = Player("Jonas", 24, 5, 13)

print(player1.name, "KDA is", player1.kills, "/", player1.deaths, "/", player1.assists)
```

## Key Definitions

1. A Class is a template for information to be stored, another example is that a Class is like a blueprint, it is something that is a guide in a sense on how the information is supposed to be held within Python.

2. An Object can be seen as the unique information that is stored following the format of the Class.

3. The function __init__ is short for initialize, it is a method that Python runs automatically whenever a new Object is created using the Class, it is used to help setup the Object's start attributes following the Classes blueprint like (name, kills, deaths).

4. The use of self can be thought of like a place holder for what ever unique Object that will be created, Python automatically runs the new Object through self in the background, as a side self can be any word but the industry standard is to use self and self must always be the first variable in any Class.

```python
class Player:
    def __init__(self, name, kills, deaths, assists):
        self.name = name
        self.kills = kills
        self.deaths = deaths
        self.assists = assists

    def kda_ratio(self):
        return self.kills / self.deaths

player1 = Player("Jonas", 21, 2, 8)

player2 = Player("Bronson", 5, 16, 2)

print(player1.name, "KDA ratio: ", player1.kda_ratio())

print(player2.name, "KDA ratio: ", player2.kda_ratio())
```

## Methods

1. A Method is defined within a Class, it uses information gained through an Object to complete an action or calculation, you can think of Methods as tools for Objects to 'do things' with their attributes.

2. A regular function lives on its own in a file and is called directly like my_function(), a Method lives inside a Class, requires self as its first parameter, and must be called on an Object like player1.kda_ratio(), Methods are tied to the Object they belong to, regular functions are not.

3. Looking at the code above, the Class simply stored that data within its preset template, while the Method kda_ratio re-interpreted the data and returned a different format using the data stored within the Class.

```python
class Player:
    def __init__(self, name, kills, deaths, assists):
        self.name = name
        self.kills = kills
        self.deaths = deaths
        self.assists = assists
    
    def kda_ratio(self):
        return self.kills / self.deaths

class ProPlayer(Player):
    def __init__(self, name, kills, deaths, assists, team, winnings):
        super().__init__(name, kills, deaths, assists)
        self.team = team
        self.winnings = winnings

pro1 = ProPlayer("Faker", 350, 80, 200, "T1", 120000)
print(pro1.name, "plays for", pro1.team, "and has won $", pro1.winnings)
print(pro1.name, "KDA ratio:", pro1.kda_ratio())
```
